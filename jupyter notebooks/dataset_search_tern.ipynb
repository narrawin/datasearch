```python
import pandas as pd, requests, json
import numpy as np
import matplotlib.pyplot as plt
from requests.utils import requote_uri
from io import StringIO

pd.set_option('display.max_columns', 500)
%matplotlib inline
```

```python
# specify filter and max results requested

searchTerm = "canopy cover"
maxHits = 5
```

```python
# TERN search and filter api endpoint
api = "https://portal.tern.org.au/search/filter/"

# Note: need double curly brackets where f string is supposed to ignore the json curly brackets
searchParams = f'{{ \"params\": {{ \"q\": \"{searchTerm}\", \"status\": \"PUBLISHED\", \"classFilter\": \"collection\", \"groupFilter\": \"All\", \"topicFilter\": \"All\", \"licenseFilter\": \"All\", \"typeFilter\": \"All\", \"subjectFilter\": \"All\", \"anzsrcforFilter\": \"All\", \"anzsrcseoFilter\": \"All\", \"gcmdFilter\": \"All\", \"temporalresFilter\": \"All\", \"verticalresFilter\": \"All\", \"horizontalresFilter\": \"All\", \"datagroupFilter\": \"All\", \"dataresolutionFilter\": \"All\", \"platformFilter\": \"All\", \"instrumentFilter\": \"All\", \"methodFilter\": \"All\", \"apniFilter\": \"All\", \"afdFilter\": \"All\", \"distributorFilter\": \"All\", \"authorFilter\": \"All\", \"datatypeFilter\": \"All\", \"ternRegionFilter\": \"All\", \"spatialPolygon\": \"\", \"temporal\": \"All\", \"sort\": \"_score desc\", \"page\": 1, \"num\":{maxHits}, \"mapsearch\": \"0\" }}}}'

headers = {"Content-Type": "application/json", "Accept": "application/json"}

searchParams

response = requests.post(api, data=searchParams, headers=headers)
data = response.json()

df = pd.DataFrame(data["json"]["hits"])

df
```

```python
# TERN search and expot api endpoint (returns csv inside "export" element of json response)
api = "https://portal.tern.org.au/search/export/"

params = f'{{ \"params\": {{ \"q\": \"{searchTerm}\", \"status\": \"PUBLISHED\", \"classFilter\": \"collection\", \"groupFilter\": \"All\", \"topicFilter\": \"All\", \"licenseFilter\": \"All\", \"typeFilter\": \"All\", \"subjectFilter\": \"All\", \"anzsrcforFilter\": \"All\", \"anzsrcseoFilter\": \"All\", \"gcmdFilter\": \"All\", \"temporalresFilter\": \"All\", \"verticalresFilter\": \"All\", \"horizontalresFilter\": \"All\", \"datagroupFilter\": \"All\", \"dataresolutionFilter\": \"All\", \"platformFilter\": \"All\", \"instrumentFilter\": \"All\", \"methodFilter\": \"All\", \"apniFilter\": \"All\", \"afdFilter\": \"All\", \"distributorFilter\": \"All\", \"authorFilter\": \"All\", \"datatypeFilter\": \"All\", \"ternRegionFilter\": \"All\", \"spatialPolygon\": \"\", \"temporal\": \"All\", \"sort\": \"_score desc\" }}}}'

response = requests.post(api, data=searchParams, headers=headers)
data = response.json()

resultCsv = StringIO(data["json"]["export"])

df = pd.read_csv(resultCsv)

df
                  
                  
```
